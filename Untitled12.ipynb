```python
def rotate_it(matrix):

    rows = len(matrix)
    columns = len(matrix[0])


    rotated_matrix = [[0] * columns for _ in range(rows)]

    for i in range(rows):
        for j in range(columns):

            new_i_value = j
            new_j_value = rows - i - 1


            rotated_matrix[new_i_value][new_j_value] = matrix[i][j]

    return rotated_matrix


matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

my_matrix = rotate_it(matrix)
for row in my_matrix:
    print(row)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger')


def count_pos_tags(text):

    words = word_tokenize(text)


    tagged_words = pos_tag(words)


    counts = {
        'nouns': 0,
        'verbs': 0,
        'pronouns': 0,
        'adjectives': 0
    }


    for word, tag in tagged_words:
        if tag.startswith('N'):
            counts['nouns'] += 1
        elif tag.startswith('V'):
            counts['verbs'] += 1
        elif tag.startswith('PRP'):
            counts['pronouns'] += 1
        elif tag.startswith('JJ'):
            counts['adjectives'] += 1

    return counts


phrase1 = "She walked quickly to the new supermarket."
phrase2 = "They swiftly jumped over the tall, wooden fence."
phrase3 = "After the long journey, the happy family finally arrived at the beautiful beach."

result1 = count_pos_tags(phrase1)
print(result1)

result2 = count_pos_tags(phrase2)
print(result2)

result3 = count_pos_tags(phrase3)
print(result3)
```

```python
import numpy as np
from keras.datasets import mnist
from keras.utils import to_categorical


(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense


model1 = Sequential()
model1.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model1.add(MaxPooling2D((2, 2)))
model1.add(Conv2D(64, (3, 3), activation='relu'))
model1.add(MaxPooling2D((2, 2)))
model1.add(Flatten())
model1.add(Dense(128, activation='relu'))
model1.add(Dense(10, activation='softmax'))


model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history1 = model1.fit(x_train, y_train, epochs=10, batch_size=128, validation_split=0.1)


_, accuracy1 = model1.evaluate(x_test, y_test)
print(f"Accuracy of Architecture 1: {accuracy1 * 100:.2f}%")
```

```python
from keras.layers import AveragePooling2D


model2 = Sequential()
model2.add(Conv2D(6, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 1)))
model2.add(AveragePooling2D(pool_size=(2, 2)))
model2.add(Conv2D(16, kernel_size=(5, 5), activation='relu'))
model2.add(AveragePooling2D(pool_size=(2, 2)))
model2.add(Flatten())
model2.add(Dense(120, activation='relu'))
model2.add(Dense(84, activation='relu'))
model2.add(Dense(10, activation='softmax'))


model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


history2 = model2.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.1)


_, accuracy2 = model2.evaluate(x_test, y_test)
print(f"Accuracy of Architecture 2: {accuracy2 * 100:.2f}%")



```

```python
import pandas as pd

data = {
    'Architecture': ['Simple CNN', 'LeNet-5'],
    'Parameters': [model1.count_params(), model2.count_params()],
    'Accuracy': [accuracy1, accuracy2 ]
}

compare_table = pd.DataFrame(data)
print(compare_table)


```
